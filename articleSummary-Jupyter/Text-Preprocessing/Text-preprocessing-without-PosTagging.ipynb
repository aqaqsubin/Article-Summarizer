```python
BASE_DIR = "/data/ksb/TestDir/sample_articles"
ORIGIN_PATH = os.path.join(BASE_DIR,"Origin-Data")
TARGET_PREPROCESSED_PATH = os.path.join(BASE_DIR,"Preprocessed-Target-Data")
TRAIN_PREPROCESSED_PATH = os.path.join(BASE_DIR,"Preprocessed-Train-Data")
SWORDS_PATH = os.path.join(BASE_DIR, "StopWordList.txt")
```

```python
class TextPreprocessor:
    def __init__(self):
        self.retokenize = RegexpTokenizer("[\w]+")
        self.swords = []
        self.tokenizer = {}

    def removeDuplicateSpace(self, text):
        return re.sub('\s+', ' ', text)  # 중복 공백, 탭, 개행 제거
    
    def removeSpecialChar(self, text):
        return ' '.join(self.retokenize.tokenize(text))

    def loadSwords(self, filename):
        self.swords = []
        with open(filename, 'r', encoding='utf-8') as f:
            swords = f.readlines()
            self.swords = [sword.strip() for sword in self.swords]

        # self.tokenizer = lambda sent: filter(lambda x:x not in self.swords and x[1] in ('NNG', 'NNP', 'VV', 'VA'), self.tagger.pos(sent))
        self.tokenizer = lambda sent: filter(lambda x:x not in self.swords, sent.split())

        return self.swords
        
    def removeSwords(self, text):
        return ' '.join([token for token in list(self.tokenizer(text))])
```

```python
def saveTextFile(baseDir, media, filename, sentences):

    mkdir_p(os.path.join(baseDir, media))
    save_path = os.path.join(os.path.join(baseDir, media), filename)

    with open(save_path, 'w', encoding='utf-8') as f:
        f.write('/n'.join([sentence for sentence in sentences if sentence is not '']))
```

```python
del_folder(TRAIN_PREPROCESSED_PATH)
del_folder(TARGET_PREPROCESSED_PATH)

preprocessor = TextPreprocessor()
preprocessor.loadSwords(SWORDS_PATH)

if __name__ == '__main__':
    
    media_list = os.listdir(ORIGIN_PATH)

    for media_idx, media in enumerate(media_list):

        media_path = os.path.join(ORIGIN_PATH, media)
        article_list = os.listdir(media_path)

        for article_idx, article_name in enumerate(article_list):
            
            reader = ArticleReader(os.path.join(media_path, article_name))
            article = Article(list(filter(None, reader)))

            trainPreprocessed = []
            targetPreprocessed = []
            for line in article.readContent() :
                cleanLine = preprocessor.removeDuplicateSpace(line)
                cleanLine = preprocessor.removeSpecialChar(cleanLine)
                
                targetLine = preprocessor.getTargetData(cleanLine)

                targetPreprocessed.append(targetLine)
                trainPreprocessed.append(cleanLine)

            print("Media : {media_name} ({media_num}/{total_media}),\
            Article : {article_name} ({article_num}/{total_article})"\
                  .format(media_name= media, media_num=media_idx, total_media=len(media_list),
                          article_name = article_name,
                          article_num = article_idx,
                          total_article=len(article_list)))
            
            saveTextFile(TARGET_PREPROCESSED_PATH, media, article_name, targetPreprocessed)
            saveTextFile(TRAIN_PREPROCESSED_PATH, media, article_name, trainPreprocessed)
```
