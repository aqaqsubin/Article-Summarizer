```python
from gensim.summarization.summarizer import summarize
from glob import iglob
import csv
import pandas as pd
import os
```

```python
BASE_DIR = "/data/ksb/TestSampleDir/articles"
PREPROCESSED_PATH = os.path.join(BASE_DIR,"Preprocessed-Data")
PRETTY_PATH = os.path.join(BASE_DIR,"Pretty-Data")
ORIGIN_PATH = os.path.join(BASE_DIR,"Origin-Data")
SUMMARY_PATH = os.path.join(BASE_DIR,"Summary-Data")
SWORDS_PATH = os.path.join(BASE_DIR, "StopWordList.txt")
```

```python
def mkdir_p(path):
    import errno
    try:
        os.makedirs(path)
    except OSError as exc:
        if exc.errno == errno.EEXIST and os.path.isdir(path):
            pass
        else:
            raise

def del_folder(path):
    try:
        rmtree(path)
    except:
        pass
```

```python
def saveCSVFile(baseDir, media, article_dist):
    save_path = os.path.join(baseDir, media) + ".csv"

    article_dist.to_csv(save_path, mode='w', header=False)
```

```python
def get_media_name(filepath):
    filename = filepath.split(os.sep)[-1]
    return filename.split(".")[0]
```

```python
del_folder(SUMMARY_PATH)
mkdir_p(SUMMARY_PATH)
```

```python
if __name__ == '__main__':
    
    entire_media = iglob(os.path.join(PREPROCESSED_PATH, '**.csv'), recursive=False)
    tmp = [os.path.join(PREPROCESSED_PATH, '중앙일보.csv')]
    
    for _, proc_article_path in enumerate(iglob(os.path.join(PREPROCESSED_PATH, '**.csv'), recursive=False)):

        media_name = get_media_name(proc_article_path)
        origin_article_path = os.path.join(PRETTY_PATH, media_name) + ".csv"
        
        print(media_name, proc_article_path)
        
        f_origin= open(origin_article_path, 'r', newline="\n", encoding="utf-8")
        summary_dist = pd.DataFrame(columns=['Origin Index', 'Title', 'Contents'])
        
        
        for idx, title, content in csv.reader(f_origin):
            text = " ".join(content.split("\t"))
            try:
                dist= {'Origin Index': idx, 'Title' : title, 'Contents' : '\t'.join(summarize(text, ratio=0.2, split=True)) }
                summary_dist = summary_dist.append(dist, ignore_index=True)
            except Exception as err:
                print(err)
                print("Drop Article : {idx}, {title}".format(idx=idx, title=title))
                pass
           
        f_origin.close()
        saveCSVFile(SUMMARY_PATH, media_name, summary_dist)   
```
